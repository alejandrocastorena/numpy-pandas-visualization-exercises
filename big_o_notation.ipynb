```python
import matplotlib.pyplot as plt
import math as math
import numpy as np
import pandas as pd
```

```python
x= (range(1,20))
y1 = [1 for n in x]
y2 = [math.log(n) for n in x]
y3 = x
y4= [(n * math.log(n)) for n in x]
y5= [n**2 for n in x]
y6= [2**n for n in x]
y7= [math.factorial(n) for n in x]
y8= [n**n for n in x]

plt.figure(figsize = (16,16))

plt.plot(x,y1, color = 'red', label= 'y1= $O(1)$')
plt.plot(x,y2, color = 'green', label= 'y2 = $O(log n)$')
plt.plot(x,y3, color = 'orange', label= 'y3= $O(n)$')
plt.plot(x,y4, color= 'cyan', label= 'y4 = $O(n log n)$')
plt.plot(x,y5, color= 'pink', label= 'y5 = $O(n^2)$')
plt.plot(x,y6, color= 'purple', label= 'y6 = $O(2^n)$',)
plt.plot(x,y7, color= 'blue', label= 'y7 = $O(n!)$')
plt.plot(x,y8, color= 'magenta', label= 'y8 = $O(n^n)$')

plt.title('Big O Notation', fontsize = 20)
plt.xlabel('Elements')
plt.ylabel('Operations')

plt.annotate('LaTex', xy= (1.25,5))

plt.legend(loc = 'upper right', fontsize = 20)
plt.xlim(1, 11)
plt.ylim(0, 50)

plt.show()
```
